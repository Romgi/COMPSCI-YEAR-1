#### Iterating Over Nested Dictionaries

```python
# Print each student's name and final grade

course = {
    "name": "COMP-SCI-1MD3",
    "students": {
        "s1": {
            "name": "Alice",
            "grades": {"midterm": 80, "final": 90}
        },
        "s2": {
            "name": "Bob",
            "grades": {"midterm": 70, "final": 85}
        },
        "s3": {
            "name": "Charlie",
            "grades": {"midterm": 95, "final": 92}
        }
    }
}

# Question:
# Using a for-loop, print each student's name and final grade
# in the format: "Alice: 90"

for student_id, info in course["students"].items():
    # print("Student ID: ", student_id)
    # print("Info: ", info)
    name = info["name"]
    final_grade = info["grades"]["final"]
    print(f"{name}: {final_grade}")

```

```python
course = {
    "name": "COMP-SCI-1MD3",
    "students": {
        "s1": {"name": "Alice", "grades": {"final": 90}},
        "s2": {"name": "Bob", "grades": {"final": 55}},
        "s3": {"name": "Charlie", "grades": {"final": 72}},
        "s4": {"name": "Dana", "grades": {"final": 48}},
    }
}

# Question:
# Suppose a student "passes" if their final grade is at least 60.
# Use the nested dictionary to count how many students passed.

pass_count = 0

for student_id, info in course["students"].items():
    final_grade = info["grades"]["final"]
    if final_grade >= 60:
        pass_count += 1

print(pass_count)  # 2


# Solution 2 without .items():

pass_count = 0

students = course["students"]   # this is a dict, e.g. {"s1": {...}, "s2": {...}, ...}

for student_id in students:     # looping over the KEYS: "s1", "s2", ...
    info = students[student_id] # get the inner dict for that student
    final_grade = info["grades"]["final"]
    if final_grade >= 60:
        pass_count += 1

print(pass_count)




```

```python
course = {
    "name": "COMP-SCI-1MD3",
    "sections": {
        "A": {
            "instructor": "Dr. Smith",
            "students": [
                {"name": "Alice", "grades": {"final": 80}},
                {"name": "Bob", "grades": {"final": 70}},
            ]
        },
        "B": {
            "instructor": "Dr. Lee",
            "students": [
                {"name": "Charlie", "grades": {"final": 90}},
                {"name": "Dana", "grades": {"final": 100}},
                {"name": "Eli", "grades": {"final": 85}},
            ]
        }
    }
}

# Question:
# For each section, compute and print the average final grade.
# Output format (order doesn't matter):
# Section A: 75.0
# Section B: 91.6666666667  (or similar)


for section_name, section_info in course["sections"].items():
    students = section_info["students"]
    total = 0
    count = 0

    for student in students:
        total += student["grades"]["final"]
        count += 1

    avg = total / count
    print(f"Section {section_name}: {avg}")

```

```python
# Compute the average assignment grade for each student

# Example output:
# Alice : 87.33333333333333

school = {
    "course_code": "COMP-SCI-1MD3",
    "students": {
        "s1": {
            "name": "Alice",
            "year": 1,
            "contact": {
                "email": "alice@mail.com",
                "phone": "555-1111"
            },
            "grades": {
                "assignments": [85, 90, 87],
                "midterm": 82,
                "final": 91
            }
        },
        "s2": {
            "name": "Bob",
            "year": 2,
            "contact": {
                "email": "bob@mail.com",
                "phone": "555-2222"
            },
            "grades": {
                "assignments": [70, 75, 72],
                "midterm": 65,
                "final": 80
            }
        },
        "s3": {
            "name": "Charlie",
            "year": 1,
            "contact": {
                "email": "charlie@mail.com",
                "phone": "555-3333"
            },
            "grades": {
                "assignments": [95, 100, 98],
                "midterm": 94,
                "final": 97
            }
        }
    }
}


# Solution:

for sid in school["students"]:
    grades = school["students"][sid]["grades"]["assignments"]
    avg = sum(grades) / len(grades)
    print(school["students"][sid]["name"], ":", avg)

```
