```python
# Solution:

def replace_at(s: str, i: int, char: str) -> str:
    """
    Replaces the character at index `i` in the string `s` with `char` using recursion.
    """
    # Base case: If `i` is 0, replace the first character and return the rest of the string
    if i == 0:
        return char + s[1:]
    
    # Recursive step: Build the result one character at a time
    return s[0] + replace_at(s[1:], i - 1, char)

# >>> replace_at("abcd", 2, "z")
# 'abzd'

# >>> replace_at("hello", 0, "H")
# 'Hello'

# >>> replace_at("world", 4, "!")
# 'worl!'

```

```python

def is_palindrome(s: str) -> bool:
    """
    Recursively checks whether the string s is a palindrome.
    """

    # Base case: empty string or 1 char is always a palindrome
    if len(s) <= 1:
        return True

    # If the first and last characters differ â†’ not a palindrome
    if s[0] != s[-1]:
        return False

    # Otherwise, check the inside
    return is_palindrome(s[1:-1])




def is_k_palindrome(s: str, k: int) -> bool:
    """
    Returns True iff s can be turned into a palindrome
    by removing up to k characters from it.
    """

    # If it's already a palindrome, no need to remove anything
    if is_palindrome(s):
        return True

    # If we have no removals left and it's not a palindrome, we fail
    if k == 0:
        return False

    # If the string is empty or 1 char, it's trivially a palindrome
    if len(s) <= 1:
        return True

    # If first and last characters match, keep them and recurse on the middle
    if s[0] == s[-1]:
        return is_k_palindrome(s[1:-1], k)

    # Otherwise, we must remove one character (using up 1 from k).
    # We try both options:
    #   - remove the first character
    #   - remove the last character
    # If either leads to a k-palindrome, we return True.
    return (
        is_k_palindrome(s[1:], k - 1) or
        is_k_palindrome(s[:-1], k - 1)
    )
```
